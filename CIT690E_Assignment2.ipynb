Q1- plot Activation function with
Linear Units, Sigmoid Units, Hyperbolic Tangent Units, and Rectified Linear
Units (ReLU), where the independent variable, x, is between -10 and 10.




```python
#import Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define sigmoid
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
#define Tanh
def tanh(x):
    return np.tanh(x)
```

```python
#define Relu
def relu(x):
  return  x * (x > 0) 
#Note:The bool result will be turned into 0 or 1 and then will get multiplied elementwise
```

```python
#define plotting function
def plotActivationFunction(interval,function,title):
    plt.plot(interval, function(interval))
    plt.title(title)
    plt.show()
```

```python
#initailaize interval
interval= np.linspace(-10, 10)
#Call Functions
plotActivationFunction(interval,sigmoid,"sigmoid")
plotActivationFunction(interval,tanh,"Tanh")
plotActivationFunction(interval,relu,"relu")
```

Q2â€“ Given an input vector x=[1,0] and output y= 1, write equations and find the activations, and the MSE error corresponding to the following feedforward neural network, where the bias is zero and the activation function
used is Sigmoid
Q2.PNG

Solution Implementation: <br>

```python
#input Matrix
input=np.array([1,0])
input.shape
```

```python
#weight Matrix
weights= np.array([[0.8,0.4, 0.3], [0.2,0.9,0.5]] )
weights.shape
```

```python
#Calculate First (a) Layer 
weights.transpose().shape
```

```python
a=weights.transpose().dot(input)
a
```

```python
#calculate h 
h=sigmoid(a)
```

```python
# initialize weights 2 output 
weights2= np.array([0.3, 0.5,0.9])
weights2.shape
```

```python
#Calculate the output
output=weights2.transpose().dot(h)
```

```python
output
```

```python
# Calculate MSE
# Mean Squared Error
def MSE(trueValue,predictedValue):
  return np.square(np.subtract(trueValue,predictedValue)).mean()
```

```python
print(MSE(1,output))
```
